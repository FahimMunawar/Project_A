```python
from tkinter import *
from PIL import ImageTk,ImageDraw
from tkinter import ttk, messagebox
import pymysql
class login:
    def __init__(self,root):
        self.root=root
        self.root.title("Login Window")
        self.root.geometry ("1350x700+0+0")
        self.root.config(bg='black')
        self.root.iconbitmap('E:\Jupyter_Codes\project\icon.ico')
        #==Bg Image===
        
        self.bg=ImageTk.PhotoImage(file="Images/low.jpg")
        bg=Label(self.root,image=self.bg).place(x=0,y=0,width=1500)
        #==Left Image===
        
        self.left=ImageTk.PhotoImage(file="Images/project.png")
        left=Label(self.root,image=self.left).place(x=200,y=150,width=500,height=438)
        
        #==Login Frame==
        
        frame1=Frame(self.root,bg='white')
        frame1.place(x=700,y=150,width=500,height=438)
        
        title=Label(frame1, text="Student Login",font=('times new roman',30,'bold'),bg='white',fg='blue').place(x=130,y=30)
        
        email=Label(frame1, text="Email",font=('times new roman',20,'bold'),bg='white',fg='grey').place(x=130,y=130)
        self.txt_email=Entry(frame1,font=('times new roman',14,),bg='lightgray')
        self.txt_email.place(x=130,y=170,width=300,height=30)
        
        password=Label(frame1, text="Password",font=('times new roman',20,'bold'),bg='white',fg='grey').place(x=130,y=200)
        self.txt_password=Entry(frame1,font=('times new roman',14,),bg='lightgray',show='*')
        self.txt_password.place(x=130,y=240,width=300,height=30)
        
        btn_forget=Button(frame1, text="Forgot Your Password?",font=('times new roman',12,),cursor='hand2',command=self.forget_password_window,bg='white',bd='0',fg='black').place(x=140,y=280)
        btn_reg=Button(frame1, text="Register New Account?",font=('times new roman',12,),cursor='hand2',command=self.register_window,bg='white',bd='0',fg='black').place(x=140,y=310)
        
        self.btn_img=ImageTk.PhotoImage(file='Images/lg2.png')
        btn_ligin=Button(frame1,image=self.btn_img,bd=0,cursor='hand2',command=self.login).place(x=130,y=350)

    def register_window(self):
        self.root.destroy()
        %run Stu_Registration.ipynb
    
    def reset(self):
        self.cmb_quest.current(0)
        self.txt_new_pass.delete(0,END)
        self.txt_answer.delete(0,END)
        self.txt_password.delete(0,END)
        self.txt_email.delete(0,END)
        
    def forget_password(self):
        if self.cmb_quest.get()=="Select" or self.txt_answer.get()=="" or self.txt_new_pass.get()=="":
            messagebox.showerror("Error",'All fields are required',parent=self.root2)
        else:
            try:
                con = pymysql.connect(host='localhost',user='root',password='',database='project')
                cur=con.cursor()
                cur.execute('select * from student_register where email=%s and question=%s and answer=%s',(self.txt_email.get(),self.cmb_quest.get(),self.txt_answer.get()))
                row=cur.fetchone()
                if row== None:
                    messagebox.showerror("Error","Please Enter the correct Security Question / Enter Answer",parent=self.root2)
                else:
                    cur.execute('update student_register set password=%s where email=%s',(self.txt_new_pass.get(),self.txt_email.get()))
                    con.commit()
                    con.close()
                    messagebox.showinfo("Success","Your Password has been updated, Please Login with new password.",parent=self.root2)                 
                    self.reset()
                    self.root2.destroy()
                    
            except Exception as es:
                messagebox.showerror("Error",f"Error due to: {str(es)}",parent=self.root)
            
    
    def forget_password_window(self):
        if self.txt_email.get()=="":
            messagebox.showerror("Error","Please Enter your email address to reset your password",parent=self.root)
        else:
            try:
                con = pymysql.connect(host='localhost',user='root',password='',database='project')
                cur=con.cursor()
                cur.execute('select * from student_register where email=%s',self.txt_email.get())
                row=cur.fetchone()
                if row== None:
                    messagebox.showerror("Error","Please Enter a valid email address to reset your password",parent=self.root)
                else:
                    con.close() 
                    self.root2=Toplevel()
                    self.root2.title("Forgot Password?")
                    self.root2.geometry("350x400+495+150")
                    self.root2.config(bg="white")
                    self.root2.focus_force()      
                    self.root2.grab_set()

                    t = Label(self.root2,text="Forget Password",font=("times new roman",20,"bold"),bg="white",fg="red").place(x=0,y=10,relwidth=1)

                    question=Label(self.root2, text="Security Question",font=('times new roman',15,'bold'),bg='white',fg='grey').place(x=50,y=100)

                    self.cmb_quest=ttk.Combobox(self.root2,font=('times new roman',13,),state='readonly',justify=CENTER)
                    self.cmb_quest['values']=("Select","Your First Pet Name","Your Birth Place","Your Best Friend Name")
                    self.cmb_quest.place(x=50,y=130,width=250)
                    self.cmb_quest.current(0)

                    answer=Label(self.root2, text="Answer",font=('times new roman',15,'bold'),bg='white',fg='grey').place(x=50,y=180)
                    self.txt_answer=Entry(self.root2,font=('times new roman',15,),bg='lightgray')
                    self.txt_answer.place(x=50,y=210,width=250)

                    new_password=Label(self.root2, text="New Password",font=('times new roman',15,'bold'),bg='white',fg='grey').place(x=50,y=260)
                    self.txt_new_pass=Entry(self.root2,font=('times new roman',15,),bg='lightgray')
                    self.txt_new_pass.place(x=50,y=290,width=250)

                    btn_change_password= Button(self.root2,text="Reset Password",command=self.forget_password, bg='green',fg='white',font=("times new roman",15,"bold")).place(x=90,y=340)
                   
                  
            except Exception as es:
                messagebox.showerror("Error",f"Error due to: {str(es)}",parent=self.root)
                
        
              
    def login(self):
        if self.txt_email.get()=="" or self.txt_password.get()=="":
            messagebox.showerror("Error","All Fields Are Required!",parent=self.root)
        else:
            try:
                con = pymysql.connect(host='localhost',user='root',password='',database='project')
                cur=con.cursor()
                cur.execute('select * from student_register where email=%s and password=%s',(self.txt_email.get(),self.txt_password.get()))
                row=cur.fetchone()
                #print(row)
                
                
                if row== None:
                    messagebox.showerror("Error","Invalid Email & Password",parent=self.root)
                    
                else:
                    
                    messagebox.showinfo("Success","Welcome",parent=self.root)
                    cur.execute('insert into temp2(email) values(%s)', (self.txt_email.get()))
                    con.commit()   
                    con.close()
                    self.root.destroy()
                    %run Student_Main.ipynb   
                con.close()   
      
                
            except Exception as es:
                messagebox.showerror("Error",f"Error due to: {str(es)}",parent=self.root)
        
        
        
        
        
        
root=Tk()
obj=login(root)
root.mainloop()
        
```
